### Testes de Hipóteses

```python
def teste_hipoteses(dataframe, target, variable1, variable2=None, alpha=0.05):
    if variable2 == None:
        control_results = dataframe[dataframe[variable1] == 1][target]
        treatment_results = dataframe[dataframe[variable1] == 0][target]
    else:
        control_results = dataframe[dataframe[variable1] == 1][target]
        treatment_results = dataframe[(dataframe[variable1] == 0) & (dataframe[variable2] == 1)][target]

    n_con = control_results.count()
    n_treat = treatment_results.count()
    successes = [control_results.sum(), treatment_results.sum()]
    nobs = [n_con, n_treat]

    z_stat, pval = proportions_ztest(successes, nobs=nobs)

    print(f'z statistic: {z_stat:.2f}')
    print(f'p-value: {pval:.3f}')
    
    if pval > alpha:
        print('NÃO podemos rejeitar a Hipótese Nula (Hₒ), ou seja, não há uma diferença entre as proporções.')
    else:
        print('REJEITAMOS a Hipótese Nula (Hₒ), ou seja, há uma diferença entre as proporções')
```

#### Learning Curves

```python
def plot_learning_curve(modelo, X, Y, CV, scoring, train_size=100):
    train_sizes, train_scores, test_scores, fit_times, _ = \
    learning_curve(estimator=modelo, X=X, y=Y, cv=CV, scoring=scoring, n_jobs=-1,
                   train_sizes=np.linspace(.1, 1.0, 100),
                   return_times=True)
    
    train_scores_mean = -np.mean(train_scores, axis=1)
    train_scores_std = -np.std(train_scores, axis=1)
    test_scores_mean = -np.mean(test_scores, axis=1)
    test_scores_std = -np.std(test_scores, axis=1)
    
    plt.figure(figsize=(20, 5))
    plt.title('Learning Curve')
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    plt.grid()
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                        train_scores_mean + train_scores_std, alpha=0.1,
                        color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                        test_scores_mean + test_scores_std, alpha=0.1,
                        color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
                label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
                label="Cross-validation score")
    plt.legend(loc="best")
    plt.show()
```
